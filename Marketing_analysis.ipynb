```python
campaign_success = df[['AcceptedCmp1','AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response']].sum() 

campaign_success
```

```python
campaign_success.plot(kind='bar')

plt.title("Campaign Acceptance")
plt.ylabel("Number Of Customers")
plt.show()
```

```python
sns.histplot(df['Income'], bins=30)

plt.title("Income Distribution")
plt.show()
```

```python
spending_cols = ['MntWines', 'MntMeatProducts', 'MntFruits', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']

df[spending_cols].sum().plot(kind='bar')

plt.title("Total Spending by Product Category")
plt.show()
```

```python
channel_cols = ['NumWebPurchases', 'NumStorePurchases', 'NumCatalogPurchases']

df[channel_cols].sum().plot(kind='bar')

plt.title("Purchases By Channel")
plt.show()
```

```python
sns.histplot(df['Recency'], bins=30)

plt.title(" Customer Recency Distribution")
plt.show()
```

```python
df['IncomeGroup'] = pd.cut(
    df['Income'],
    bins=[0,30_000,70_000,200_000],
    labels=['Low', 'Middle', 'High']
)
```

```python
pd.crosstab(df['IncomeGroup'],df['Response'])
```
